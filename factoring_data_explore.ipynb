###  Загружаем файл с данными по балансам компаний
датафрейм занимает около 5 Гб

```python
# читаем только nrows, т.к. без ограничения будет 3 млн. записей на 5 Гб
nrows = 100000
data_balance = pd.read_csv('data/data_balance.csv', 
                           sep=';', 
                           header = None, 
                           nrows=nrows
                          )
```

```python
data_balance.info()
```

```python
# Заголовок файла с балансом считываем из файла json
# '1':'Наименование', '2':'ОКПО', '3':'ОКОПФ', '4':'ОКФС', '5':'ОКВЭД', '6':'ИНН', '7':'Код единицы измерения', '8':'Тип отчета', 
import json
with open('data/balance_header.json', 'r') as file:
    balance_header = json.load(file)
```

```python
# переименовываем колонки
data_balance.columns = list(map(str, range(1,267)))
data_balance.rename(columns=balance_header, inplace=True)
```
